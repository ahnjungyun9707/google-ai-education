## 텐서의 생성

```python
import torch
#torchArray = torch.tensor([1,2,3,4])
torchArray = torch.FloatTensor([1,2,3,4])
print(torchArray)
print(type(torchArray))
print(torchArray.dtype)
print(torchArray.device)
print(torchArray.ndim)
print(torchArray.shape)
print(torchArray.size())
```

```python
import numpy as np

nparray = np.array([1,2,3,4], dtype=np.float32)
print(nparray)
print(type(nparray))
```

### 난수 생성

```python
import torch

# 1행 2열짜리 난수를 생성해준다.(0~1사이의 난수를 생성)
tensor = torch.rand(1,10)
print(tensor)
print(tensor.shape)
print(tensor.dtype)
print(tensor.device)
```

## 차원 변환

```python
import torch
#shape(1,2) 1행 2열
tensor = torch.rand(1,2)
print(tensor)
print(tensor.shape)
#reshape(2,1) 2행 1열
tensor = tensor.reshape(2,1)
print(tensor)
print(tensor.shape)
```

## 텐서 생성시 자료형 설정

## 장치 설정
CPU에서 신경망을 구성할지, GPU에서 구성할지 설정

```python
import torch

# 현재 torch가 사용할 수 있는 장치를 확인
# if torch.cuda.is_available() -> nvidia GPU를 사용할 수 있다면
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
cpu = torch.tensor([1,2,3])
gpu = torch.tensor([1,2,3], device=device)
print(cpu)
print(gpu)
print(device)
```

## 장치를 GPU로 전환

```python
# cpu -> gpu로 전환
gpu1 = cpu.cuda()
print(gpu1)
print(gpu1.device)
```

## numpy -> Tensor로 변환

```python
import torch

import numpy as np

# 기존에 읽어온 numpy ndarray가 있다고 가정
npArray = np.array([1,2,3,4])
# numpy -> tensor
tensor = torch.from_numpy(npArray)
print(tensor)
print(type(tensor))

# tensor -> numpy
npArray1 = tensor.numpy()
print(npArray1)
print(type(npArray1))
```

## 장치변환 + 넘파이변환

```python
import torch

# tensor[1,2,3,4]를 생성 하는데 GPU에 생성한다.
tensor = torch.tensor([1,2,3,4], device=torch.device("cuda:0"))

# cpu로 전환 + numpy로 변환
npArray = tensor.cpu().numpy()
print(tensor)
print(tensor.device)
```
