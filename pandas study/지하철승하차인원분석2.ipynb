```python
import pandas as pd
```

```python
csvPath ='/content/drive/MyDrive/CARD_SUBWAY_MONTH_202405.csv'
csvPath2 = '/content/drive/MyDrive/CARD_SUBWAY_MONTH_202406.csv'
csvPath3 = '/content/drive/MyDrive/CARD_SUBWAY_MONTH_202407.csv'
```

```python
df1 = pd.read_csv(csvPath, index_col=False)
df2 = pd.read_csv(csvPath2, index_col=False)
df3 = pd.read_csv(csvPath3, index_col=False)
```

```python
df1.head()
```

```python
df_1 = df1.drop(['등록일자'], axis=1)
df_2 = df2.drop(['등록일자'], axis=1)
df_3 = df3.drop(['등록일자'], axis=1)
```

```python
df_1.head()
```

```python
df_2.head()
```

```python
df_3.head()
```

```python
df_1['총승객수'] = df_1['승차총승객수'] + df_1['하차총승객수']
df_1
```

```python
df_2['총승객수'] = df_2['승차총승객수'] + df_2['하차총승객수']
df_2
```

```python
df_3['총승객수'] = df_3['승차총승객수'] + df_3['하차총승객수']
df_3
```

```python
df = pd.concat([df_1, df_2, df_3], ignore_index=True)
df
```

```python
df_m = df.groupby('역명')['총승객수'].mean()
df_m
```

```python
df_m.sort_values(ascending=False).head(5)
```

```python
df_m.sort_values(ascending=True).head(5)
```
