```python
import numpy as np
import pandas as pd
```

```python
csvPath = '/content/drive/MyDrive/등록인구_20240812141124.csv'
```

```python
pop_seoul=pd.read_csv(csvPath, encoding='UTF-8')
pop_seoul.head()
```

```python
pop_seoul.tail()
```

```python
pop_seoul.columns = pop_seoul.loc[0]
```

```python
pop_seoul.head()
```

```python
# pop_seoul의 0번행 삭제
pop_seoul = pop_seoul.drop(0)
pop_seoul.head()
```

```python
# 불필요한 컬럼 삭제
# iloc를 이용해서 컬럼 삭제
# [0,2,4,5,7,8,10,11,12]

pop_seoul = pop_seoul.iloc[:, [1,3,6,9,13]]
pop_seoul.head()
```

```python
pop_seoul.columns = ['구별','인구','한국인','외국인','고령자']
pop_seoul.head()
```

```python
# 1,2행 삭제
pop_seoul = pop_seoul.drop([1,2])
pop_seoul.head()
```

```python
# 인덱스 값을 초기화
pop_seoul = pop_seoul.reset_index(drop=True)
pop_seoul.head()
```

```python
pop_seoul.tail()
```

```python
xlsxPath = '/content/drive/MyDrive/서울시 자치구 (연도별) CCTV 설치현황_240630.xlsx'
cctv_seoul = pd.read_excel(xlsxPath, header=2, usecols='C,D,E,F,G,H,I,J,K,L,M,N')
cctv_seoul.head()
```

```python
cctv_seoul.tail()
```

```python
# 0번행 삭제
cctv_seoul = cctv_seoul.drop(0)
cctv_seoul.head()
```

```python
cctv_seoul.tail()
```

```python
# index 초기화
cctv_seoul = cctv_seoul.reset_index(drop=True)
cctv_seoul.head()
```

```python
cctv_seoul.columns = ['구별', 'CCTV대수', '2016년이전CCTV ', '2016년', '2017년', '2018년', '2019년',
       '2020년', '2021년', '2022년', '2023년', '2024년2분기']

cctv_seoul.head()
```

```python
# pop_seoul, cctv_seoul 데이터프레임 합치기
pop_cctv = pd.merge(pop_seoul, cctv_seoul, on='구별')
pop_cctv.head()
```

```python
xlsxPath2 ='/content/drive/MyDrive/서울시 자치구 (목적별) CCTV 설치현황_240630.xlsx'
```

```python
cctv_seoul2 = pd.read_excel(xlsxPath2, header=3, usecols='B,D,E,F,G,H')
cctv_seoul2.head()
```
