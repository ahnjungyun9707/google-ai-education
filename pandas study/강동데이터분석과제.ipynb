```python
import pandas as pd

students = ['학생1', '학생2', '학생3', '학생4', '학생5']
subjects = ['국어', '영어', '수학', '과학', '사회']

data = {
    '국어': [85, 78, 75, 80, 95],
    '영어': [88, 92, 78, 85, 84],
    '수학': [100, 67, 78, 82, 76],
    '과학': [82, 95, 67, 89, 84],
    '사회': [79, 85, 87, 90, 93],
    '학생': students
}

df = pd.DataFrame(data)
df
```

```python
df = df.set_index('학생')
df
```

```python
df.mean()
```

```python
df['총점'] = df.sum(axis=1)
df['평균'] = df.mean(axis=1)
df
```

```python
# 지피티 도움
result = {}

for subject in subjects:
    top_student = df[subject].idxmax()
    top_score = df[subject].max()
    bottom_student = df[subject].idxmin()
    bottom_score = df[subject].min()
    result[subject] = [top_student, top_score, bottom_student, bottom_score]

result_df = pd.DataFrame(result, index=['최고점 학생', '최고점', '최저점 학생', '최저점'])

result_df
```
