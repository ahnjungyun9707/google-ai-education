```python
import numpy as np

# 선형회기 모델
from sklearn.linear_model import LinearRegression
```

```python
# 데이터 준비
# y = wx + b
n=100
x=np.arange(n).reshape(-1,1)

# 100행, 1열
x.shape

np.random.seed(14)
randValue = np.random.random(n)*35
randValue = randValue.reshape(-1,1)
randValue.shape

# w=4, b=7
y = 4*x + 7
y.shape
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.show()
```

```python
# 8:2
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=123)

# 모델생성
model = LinearRegression()

# 모델 학습
# x,y 훈련데이터를 fit함수에 입력
model.fit(x_train,y_train)
```

```python
# w, b을 확인
model.coef_, model.intercept_
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
y_pred = model.predict(x_test)
print(mean_squared_error(y_test, y_pred))
print(mean_absolute_error(y_test, y_pred))
print(np.sqrt(mean_squared_error(y_test, y_pred)))
```

```python
# 다항 방정식
# y = w1*x1 + w2*x2 +b
# w1=4, w2=5 , b=3
# y = 4*x1 + 5*x2 + 3
n=10
x1 = np.arange(n).reshape(-1,1)
x2 = np.arange(n).reshape(-1,1)

np.random.seed(14)
randValue =np.random.random(n)*5
randValue = randValue.reshape(-1,1)

y = 4*x1 + 5*x2 + 3 + randValue
y
```

```python
x = np.hstack((x1,x2))
x
```

```python
# 8:2
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=123)

# 모델생성
model = LinearRegression()

# 모델 학습
# x,y 훈련데이터를 fit함수에 입력
model.fit(x_train,y_train)
```

```python
# x = np.hstack((x1,x2))
# x
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
y_pred = model.predict(x_test)
print(mean_squared_error(y_test, y_pred))
print(mean_absolute_error(y_test, y_pred))
print(np.sqrt(mean_squared_error(y_test, y_pred)))
```

```python
# 수식에 3제곱을 구현하고 모델링하기
n = 100

randNum = np.random.rand(n, 1)
x = 6*randNum - 3

np.random.seed(14)
randValue =np.random.random(n)*20
randValue = randValue.reshape(-1,1)

y = 2*x**3 + x + 2 + randValue
plt.scatter(x,y)
plt.show()
```

```python
# 고차 방정식을 만들어주는 함수
from sklearn.preprocessing import PolynomialFeatures

poly_features = PolynomialFeatures(degree=4, include_bias=True)
x_poly = poly_features.fit_transform(x)
x[0], x_poly[0]
```

```python
model = LinearRegression()
model.fit(x_poly,y)
```

```python
model.coef_, model.intercept_
```

```python
xx = np.linspace(-3,3,100)
y_pred = model.predict(poly_features.fit_transform(xx.reshape(-1,1)))
plt.scatter(x,y)
plt.plot(xx,y_pred,)
plt.show()
```
