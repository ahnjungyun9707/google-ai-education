```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import LinearRegression
```

```python
n = 10
x = np.arange(n)
print(x)
print(type(x))
print(x.shape)
```

```python
x = np.arange(n).reshape(-1,1)
print(x)
print(x.shape)
```

```python
y = (2*x + 1)
```

```python
print(y)
print(y.shape)
```

```python
# numpy의 random.seed()함수에 seed값 설정
np.random.seed(14)

# 10개의 난수를 생성
# 0~1사이의 난수가 생성
randValue = np.random.random(n)*3
print(randValue)

randValue = randValue.reshape(-1,1)
y = (2*x + 1)

# 2x+1 + 난수
y = y + randValue
y
```

```python
# matplotlib으로 그려보기
import matplotlib.pyplot as plt
%matplotlib inline
plt.scatter(x,y)
plt.show()
```

```python
# STEP 1. 모델 선언
model = LinearRegression()
```

```python
# STEP 2. 학습
# x데이터(문제)와 y데이터(레이블) -> 지도학습
model.fit(x, y)
```

```python
# 모델에서 찾은 기울기(w)
print(model.coef_)
# 모델에서 찾은 바이어스(b)
print(model.intercept_)
```

```python
# STEP 3. 예측
prediction = model.predict([[20.0],[21.0]])
```

```python
prediction
```
