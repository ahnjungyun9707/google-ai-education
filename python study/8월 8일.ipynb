```python
#1. 데이터 불러오기
df = pd.read_csv(csvPath, encoding = 'CP949')

# 2. 결측치 있는 행 삭제 (4~12월 데이터가 없는 행 삭제)
df.dropna(how='any', inplace=True)

def yearTotal(year):
    # 특정 해의 데이터만 선택해서 데이터프레임 만들기
    # 물리적으로 다른 데이터 저장공간 df_year를 생성해서 아래 코드를 처리
    df_year = df.loc[df['연도'] == year].copy()

    # 관광지별 모든 1~12월 방문객수를 더한 총방문객수 계산
    df_year['총방문객수'] = df_year.loc[:,'1월':'12월'].sum(axis=1)

    # 2010년의 관광지별 방문객수 모두 더하기
    total = int(df_year['총방문객수'].sum())  # 7183542.0
    return total

for year in range(2010, 2022):
    print(f'{year}년 : {yearTotal(year)}')
```

```python
def monthTotal(year, month):
    # 입력된 연도에 해당하는 데이터만 필터링
    df_year = df.loc[df['연도'] == year].copy()

    # 입력된 달의 총 방문객 수를 계산
    total = int(df_year[month].sum())
    return total

# 2010년부터 2021년까지 각 연도별 각 달의 총 방문객 수를 출력
months = ['1월', '2월', '3월', '4월', '5월', '6월', '7월', '8월', '9월', '10월', '11월', '12월']
for year in range(2010, 2022):
    for month in months:
        print(f'{year}년 {month}: {monthTotal(year, month)}')
```

```python
def monthWithMostVisitors(year):
    df_year = df.loc[df['연도'] == year].copy()

    month_visitors = {month: int(df_year[month].sum()) for month in months}
    max_month = max(month_visitors, key=month_visitors.get)
    max_visitors = month_visitors[max_month]

    return max_month, max_visitors

months = ['1월', '2월', '3월', '4월', '5월', '6월', '7월', '8월', '9월', '10월', '11월', '12월']
for year in range(2010, 2022):
    max_month, max_visitors = monthWithMostVisitors(year)
    print(f'{year}년 가장 방문객이 많은 달: {max_month}, 방문객 수: {max_visitors}')
```
