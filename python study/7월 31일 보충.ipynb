```python
# 숫자 n이 있을때 1~n까지 숫자 중 3의 배수만 출력

n=15

for i in range(1, n+1):
    if i % 3 == 0: # i이가 3의 배수인가요? -> i를 3으로 나누면 나머지가 0인가요?
        print(i, end = ' ')

for i in range(3, n+1, 3):
    print(i, end = ' ')
```

```python
# 숫자 n이 있을때 1~n까지 숫자 중 3의 배수만 출력

n=15
a=3
for i in range(1, n+1):
    if i % a == 0: # i이가 a의 배수인가요? -> i를 a으로 나누면 나머지가 0인가요?
        print(i, end = ' ')

for i in range(a, n+1, a):
    print(i, end = ' ')
```

```python
# 숫자 n이 있을때 1~n까지 숫자 중 a의 배수랑  b의배수 출력

n=15
a=3
b=5


for i in range(1, n+1):
    if i % a == 0 or i % b == 0: # i이가 a의 배수인가요? -> i를 a으로 나누면 나머지가 0인가요?
        print(i, end = ' ')
```

```python
li=[3,4,5,1,2,7,8,9,6,8]

# 5 이상의 수가 있는 위치만 출력

for i in li:
   if i >= 5:
    print(i, end = ' ') # 인덱스 정보를 제외하고 list 자체를 for문과 활용

print()

for i in range(len(li)):
  if li[i] >=5:
        print(f'{(li[i])} 숫자가 {i}번에 위치하고 있습니다.' ) # 인덱스를 활용해서 for문과 list를 다룬다.
```
