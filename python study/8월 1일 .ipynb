```python
import numpy as np

def softmax(X):
    exp_a = np.exp(X)
    sum_exp_a = np.sum(exp_a) # This line was incorrectly indented
    y = exp_a / sum_exp_a

    return y
```

```python
testList = [10,7 ,4,2,2]
result = softmax(testList)
print(result)
```

```python
# 팰린드롬 문자열
# 영어로 팰리드롬인 문자열을 알려줘
# 문자열의 갯수가 홀수인지 짝수인지 확인
```

```python
def palindrom():
  myStr = input("문자열을 입력하세요.:")
  print(len(myStr))

  palin = True
  for i in range(len(myStr)//2):
      if myStr[i] != myStr[-(i+1)]:
          print("팰린드롬이 아닙니다")
          palin = False
          break
  else:
    print("팰린드롬입니다")

    return palin
```

```python
result = palindrom()
print(result)
```
